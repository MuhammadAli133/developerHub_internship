```python
# Task 1: Exploring and Visualizing a Simple Dataset

## Problem Statement
#The goal of this task is to explore and visualize a dataset to understand its structure, trends, and patterns.

## Dataset
#We will use the Iris dataset, which contains measurements of flowers such as sepal length, sepal width, petal length, and petal width, along with their species classification.

## Objective
#- Load and inspect the dataset
#- Perform summary statistics
```

```python
# Import necessary libraries
import pandas as pd              # For data handling
import numpy as np               # For numerical operations
import matplotlib.pyplot as plt # For plotting
import seaborn as sns           # For advanced visualization

# Load the Iris dataset from seaborn
df = sns.load_dataset('iris')

# Display the first 5 rows of the dataset
df.head()
```

```python
# Print the shape of the dataset (rows, columns)
print("Shape of dataset:", df.shape)

```

```python
# Print column names
print("\nColumn Names:")
print(df.columns)

```

```python
# Display first 5 rows again (for clarity)
print("\nFirst 5 Rows:")
print(df.head())

```

```python
# Get dataset information (data types, null values)
print("\nDataset Info:")
df.info()

```

```python
# Get statistical summary of numerical columns
print("\nStatistical Summary:")
print(df.describe())
```

```python
# Set style for better visuals
sns.set(style="whitegrid")

# Create a scatter plot
plt.figure(figsize=(8,6))

sns.scatterplot(
    x='sepal_length',
    y='sepal_width',
    hue='species',   # color based on species
    data=df
)

# Add title and labels
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

# Show plot
plt.show()
```

```python
# Create histograms for all numerical features
df.hist(figsize=(10,8), bins=20)

# Add a main title
plt.suptitle("Histograms of Iris Dataset Features", fontsize=16)

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

```python
# Create box plots for each feature
plt.figure(figsize=(10,6))

sns.boxplot(data=df)

# Add title
plt.title("Box Plot of Iris Dataset Features")

# Show plot
plt.show()
```

```python
# Final Insights from Iris Dataset Analysis

## 1. Dataset Overview
###- Sepal length and width show partial separation among species.
#- One species (Setosa) is clearly separable from others.

## 3. Data Distribution (Histograms)
#- Petal length and petal width show strong variation between species.
#- Sepal features are more overlapping in distribution.

## 4. Outlier Detection (Box Plot)
#- Few or no extreme outliers are present in the dataset.
#- Some variation exists in sepal width.

## 5. Conclusion
#- Petal features are more useful for classification compared to sepal features.
#- The dataset is well-structured and suitable for machine learning tasks.
```
